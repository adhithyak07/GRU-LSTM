```python
import tensorflow as tf # DeepLearning Framework
from tensorflow import keras # API for building and training deep learning models
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer # Convert text to numbers
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Download the dataset

path=tf.keras.utils.get_file(
    'shakespeare.txt',
     'https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt'
)

# Read the dataset

text=open(path,'r',encoding='utf-8').read()

# Convert text into list of sentences

text=text.lower()

```

```python
# Total characters

print("Total characters:",len(text))

words=text.split()
# Total words
print("Total words:",len(words))
```

```python
tokenizer = Tokenizer() # create a tokenizer object

tokenizer.fit_on_texts([text]) # Learn vocabulary

word_index = tokenizer.word_index # Get the word index dictionary

vocab_size = len(word_index) + 1 # Total unique words
print(" Vocabulary size:",vocab_size)
```

```python
# last 10 words

for word,index in list(word_index.items())[:10]:
    print(word,":",index)


```

```python
# convert text into sequences

sequences = tokenizer.texts_to_sequences([text])[0]

print(sequences[:30])
```

```python
# Split text into lines

lines= text.split('\n')

input_sequences=[] # Store sequenes of words

for line in lines:
    # convert line into sequence
    token_list=tokenizer.texts_to_sequences([line])[0]

    # Create n-grams
    for i in range(1,len(token_list)):
        n_gram_sequence=token_list[:i+1]
        input_sequences.append(n_gram_sequence)

print("Total sequences:",len(input_sequences))
```

### Part-2 Padding ---> Train/Test ---> LSTM Model ---> Training

```python
# Find the maximum sequence length
max_Sequence_len = max([len(x) for x in input_sequences])
print(" Maximum sequence length :",max_Sequence_len)

# Apply padding

input_sequences = np.array(
    pad_sequences(
        input_sequences,
        maxlen=max_Sequence_len,
        padding='pre'
    )
)

print("sequences & words per sequence:",input_sequences.shape)
```

```python
# Split into x & y

x=input_sequences[:,:-1]

y=input_sequences[:,-1]

print("x_shape:",x.shape)

print("y shape:",y.shape)
```

```python
# Convert o/p into categorical

y=keras.utils.to_categorical(
    y,
    num_classes = vocab_size
)
```
